```python
# 5. Show the total spent by customers from low to high.
query = """
SELECT
	c.customer_id AS customer_id,
	SUM(od.order_sales) AS total_sales
FROM customers c
	JOIN orders o
	on c.customer_id = o.customer_id
JOIN order_details od
	on o.order_id = od.order_id
GROUP BY c.customer_id
ORDER BY total_sales ASC
;
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 6. What is the most profitable city in the State of Tennessee?
query = """
SELECT shipping_city
FROM orders join order_details USING(order_id)
WHERE shipping_state = 'Tennessee'
GROUP BY shipping_city
ORDER BY SUM(order_profits) DESC
LIMIT 1
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 7. What’s the average annual profit for that city across all years?

query = """
SELECT AVG(order_profits) AS avg_annual_profit_Lebanon
FROM orders
JOIN order_details
USING (order_id)
WHERE shipping_city='Lebanon' AND shipping_state='Tennessee'

"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 8. What is the distribution of customer types in the data?
query = """
SELECT
customer_segment,
COUNT(customer_id)
FROM customers
GROUP BY customer_segment
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 9. What’s the most profitable product category on average in Iowa across all years?
query = """
SELECT DISTINCT product_category,
AVG(order_profits) AS avg_most_profitable
FROM product p
JOIN order_details od ON p.product_id=od.product_id
JOIN orders o ON o.order_id=od.order_id
WHERE shipping_state = 'Iowa'
GROUP BY product_category
ORDER BY avg_most_profitable desc
LIMIT 1
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 10. What is the most popular product in that category across all states in 2016?

query = """
SELECT product_name, SUM(quantity) AS quantity_product_sold
FROM orders o
JOIN order_details od
USING (order_id)
JOIN product p
USING(product_id)
WHERE product_category='Furniture' AND EXTRACT(YEAR FROM order_date)::Integer=2016
GROUP BY product_name
ORDER BY quantity_product_sold DESC
LIMIT 1
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 11. Which customer got the most discount in the data? (in total amount)
query = """
SELECT
    o.customer_id,
    c.customer_name,
    ROUND(CAST(SUM((od.order_sales / (1 - od.order_discount)) - od.order_sales) AS NUMERIC), 2) AS total_discount
FROM
    orders o
JOIN
    order_details od ON o.order_id = od.order_id
JOIN
    customers c ON o.customer_id = c.customer_id
GROUP BY
    o.customer_id, c.customer_name
ORDER BY
    total_discount DESC
LIMIT 1;
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 12. How widely did monthly profits vary in 2018?
query = """
with months_total AS (
  SELECT TO_CHAR(order_date, 'MM-YYYY') AS month,
  SUM(order_profits) AS month_total
  FROM orders JOIN order_details USING(order_id)
  WHERE EXTRACT(YEAR FROM order_date) = 2018
  GROUP BY TO_CHAR(order_date, 'MM-YYYY')
  )


  SELECT month,
  month_total,
  month_total - LAG(month_total) OVER(ORDER BY month) AS month_diff
  FROM months_total;

"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 13. Which was the biggest order regarding sales in 2015?
query = """
SELECT order_id, SUM(order_sales) AS total_order_sales
FROM order_details
LEFT JOIN orders
USING(order_id)
WHERE EXTRACT(YEAR FROM shipping_date)=2015
GROUP BY order_id
ORDER BY total_order_sales DESC
LIMIT 1
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 14. What was the rank of each city in the East region in 2015 in quantity?
query = """
SELECT
    o.shipping_city,
    SUM(od.quantity) AS total_quantity,
    RANK() OVER (ORDER BY SUM(od.quantity) DESC) AS rank
FROM
    orders o
JOIN
    order_details od ON o.order_id = od.order_id
WHERE
    o.shipping_region = 'East' AND
    EXTRACT(YEAR FROM o.order_date) = 2015
GROUP BY
    o.shipping_city
ORDER BY
    rank;
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 15. Display customer names for customers who are in the segment ‘Consumer’ or ‘Corporate.’ How many customers are there in total?
query = """
SELECT customer_name,
COUNT(*) OVER() as total_number_customers
FROM customers
WHERE customer_segment IN ('Consumer', 'Corporate')
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 16. Calculate the difference between the largest and smallest order quantities for product id ‘100.’

query = """
SELECT MAX(quantity)-MIN(quantity) AS difference_order_quantities
FROM order_details
WHERE product_id=100

"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 17. Calculate the percent of products that are within the category ‘Furniture.’
query = """
SELECT
    ROUND(
        (COUNT(CASE WHEN product_category = 'Furniture' THEN 1 END) * 100.0) / COUNT(*), 2
    ) AS furniture_percent
FROM
    product
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 18. Display the number of product manufacturers with more than 1 product in the product table.
query = """
SELECT product_manufacturer, COUNT(DISTINCT product_id) AS num_of_duplicates
FROM product
GROUP BY product_manufacturer
HAVING COUNT(DISTINCT product_id) > 1
ORDER BY COUNT(DISTINCT product_id)
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 19. Show the product_subcategory and the total number of products in the subcategory.
      #Show the order for the most to least number of products.
query = """
SELECT product_subcategory, COUNT(product_name) AS num_product_subcategory
FROM product
GROUP BY product_subcategory
ORDER BY num_product_subcategory DESC
"""

display(pd.read_sql(sa.text(query),connection))
```

```python
# 20. Show the product_id(s), the sum of quantities, where for each sale of product quantities is greater than or equal to 100.
query = """
SELECT
    product_id,
    SUM(quantity) AS total_quantity
FROM
    order_details
WHERE
    quantity >= 100
GROUP BY
    product_id
ORDER BY
    total_quantity DESC;
"""

display(pd.read_sql(sa.text(query),connection))
```
